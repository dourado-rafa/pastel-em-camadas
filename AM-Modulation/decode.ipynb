```python
from module import calcFFT, generateSin, low_pass_filter
import numpy as np
import sounddevice as sd
from scipy.io import wavfile as sf
import matplotlib.pyplot as plt
```

```python
duration = 5 
f_carrier = 14000
samplerate = 44100
```

```python
samplerate, signal = sf.read('audios/encoded.wav') # Lendo o arquivo WAV channel1

_, carrier = generateSin(f_carrier, duration, samplerate)
time = np.linspace(0, duration, samplerate*duration)

demoduled = carrier*signal
filtred = low_pass_filter(demoduled)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(time, demoduled)
plt.title('Audio Demodulado')

freqs, ampls = calcFFT(demoduled, samplerate)
plt.subplot(122)
plt.plot(freqs, np.abs(ampls))
plt.title('Fourier do Audio Demodulado')

plt.show()
```

```python
plt.plot(time, filtred)
plt.title('Audio Demodulado Filtrado')
plt.plot()
```

```python
sd.play(filtred, samplerate)
sd.wait()
```
